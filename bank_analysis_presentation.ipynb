# Summary And Recommendations

- Middle-income earners make up 44% of our customers, with high-income and low-income groups at 33% and 22%, respectively. This makes the middle-income segment our largest customer base.
- Spending patterns across different product categories are similar across all income classes. Mortgages and loans consistently rank as the most expensive transactions on average.
- Credit scores are also fairly evenly distributed across all income segments.
- Overall, there’s no significant difference in spending behavior or product preferences across income groups.
- However, since middle-income earners make up the majority of our customer base, products and recommendations should be tailored primarily to them.
- Based on the even distribution of product preference across all customer segments, and a dominance of middle-class income earners, you could personalize market campages to attract middle income earners
- Mortgages and loans are also the customers most expensive transactions, so create optimized product and value delivery for mortgages and loans


# Customer Segmentation Breakdown

```python
counts = df['CustomerSegment'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(counts, labels=counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Customer Segment pie")
plt.axis('equal')
plt.show
```

the above chart tells
- Middle Income earners are our most customers, making 44% of customer base followed by high income earners and low income earners

```python
# customer segment
sns.boxplot(x="CustomerSegment", y="Amount", data=df)
plt.title("Distribution of Transaction Amount across Customer Segment")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Transactions across Product subcategory and Customer Segment are eqully distributed

Their average range falling between
2k and 7k for Product Subcategory and
2k and 8k across all income earners

### Customer Segment by Product category and subcategory distribution

```python
df.groupby('CustomerSegment')['ProductCategory'].value_counts()
```

```python
counts = df.groupby('CustomerSegment')['ProductCategory'].value_counts()
counts.plot(kind='bar', figsize=(10, 6))
plt.title('Product Category by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Count')
plt.legend(title='Product Category')
plt.tight_layout()
plt.show()
```

```python
counts = df.groupby('CustomerSegment')['ProductSubcategory'].value_counts()
counts.plot(kind='bar', figsize=(10, 6))
plt.title('Product Subcategory by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Count')
plt.legend(title='Product Category')
plt.tight_layout()
plt.show()
```

The plot tells us that Middle class earners make a majority of the customers which is also clearly seen from the Customer Segment pie chart above and also there isn't much prefernce of a particular product category or subcategory over others.

```python
df.groupby('RecommendedOffer')['CustomerSegment'].value_counts()
```

Recommended Offers are products recommended based on Income class which clearly fits the present needs for each income class

```python
sns.histplot(
    data=df,
    x='CustomerScore',
    hue='CustomerSegment',
    bins=30,
    stat='count',
    common_norm='False',
    )

plt.title('Customer Credit Score Distribution by Customer Segment')
plt.xlabel('Customer Credit Score')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
df.groupby('CustomerSegment')['CustomerScore'].describe()
```

The average credit score across all income class is the same, so no class has a higher or lower credit score that usual

```python
df.groupby(['CustomerSegment', 'ProductCategory'])['Amount'].describe()
```

```python
pivot_mean = df.pivot_table(
    index='CustomerSegment',
    columns='ProductCategory',
    values='Amount',
    aggfunc='mean',
    fill_value=0
)
pivot_mean
```
